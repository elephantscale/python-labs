```python
import mysql.connector
```

```python
mydb = mysql.connector.connect(
 host="localhost",
    user="root",
    password="passNew@@123",
    database="finalDb"
)

print (mydb)
myCursor = mydb.cursor()
```

```python
## print("Press 1 to add a new customer\nPress 2 to get customers info")


def newCustomer(inputName,inputAddress):
    if len(inputName) != 0 and len(inputAddress) !=0 :           
        myCursor.execute("insert into customers (name,address) values (%s,%s)" , [inputName, inputAddress] )
        mydb.commit()
        print("A new customer has been created")     
    else:
        print("Invalid input\nName and address cannot be empty")
    
def getAllCustomers():
    myCursor.execute("select * from customers")
    myResult = myCursor.fetchall()
    for x in myResult:
        print (x)
        print("------")

def getFilteredCustomers():
    minId =int( input("Enter Min ID: ") )        
    myCursor.execute("select * from customers where id >=%s ", [minId] )
    myResult = myCursor.fetchall()
    for x in myResult:
        print (x)
        print("------")

def checkCustomerById( cid ):
    myCursor.execute("select count(*) from customers where id=%s", [ cid ] )
    myResult = myCursor.fetchone()
    return bool(myResult[0])
  
def getCustomerInfoById( cid ):
    myCursor.execute("select * from customers where id=%s", [ cid ] )
    myResult = myCursor.fetchone()
    return { "name": myResult[1], "address":myResult[2] }    
        
try:
    choice = int( input("Enter your choice: ") )        
    if choice == 1:
        inputName = input("Enter Name: ")
        inputAddress= input("Enter Address: ")
        newCustomer(inputName,inputAddress)
                    
    elif choice == 2:
        print("Press A to print all customers\nPress B to create a condition")
        print_choice = input("Enter your choice:")
        
        if print_choice.upper() == "A":
            getAllCustomers()

        elif print_choice.upper() =="B":            
           getFilteredCustomers()
      
        else:
            print("Invalid input\nPlease choose from [A,B]")
    
    elif choice == 3:
        cid = int( input("Please enter a customer ID: "))
        
        if checkCustomerById( cid ):                
            customerInfo = getCustomerInfoById( cid )
            print( "Are you sure that you want to delete a customer with name of: %s located at %s?" % (customerInfo['name'], customerInfo['address']   ) )
            deleteConfirm = input("Yes/No")

            if deleteConfirm.upper() == "YES" or deleteConfirm.upper() == "Y":            
                myCursor.execute("delete from customers where id = %s", [ cid ] )
                mydb.commit()      
                print ("Cusotmer has been deleted")
            else:
                print ("Cancelled")
                
        else:
            print("A customer with the provided ID is not available in our database")
       
    elif choice == 4:
        cid = int ( input("Enter customer ID:") )
        
        if checkCustomerById( cid ):
            customerInfo = getCustomerInfoById( cid )
            print( "Selected user,  name: %s located at %s?" % (customerInfo['name'], customerInfo['address']   ) )
            action = input("What do you want to update, n for name, a for address: ")
            
            if action.upper() == 'N':
                newName = input("Please enter new name: " )
                myCursor.execute("update customers set name = %s where id = %s" , [ newName,cid  ])
                mydb.commit()                
                
            elif action.upper() == "a":
                newAddr = input("Please enter new address: " )
                myCursor.execute("update customers set address = %s where id = %s" , [ newAddr,cid  ])
                mydb.commit()
                
            else:
                print("Invalid input\nPlease choose from [A,N]")
        else:
            print("A customer with the provided ID is not available in our database")
    else:
        print("Invalid input\nPlease choose from [1,2]")

except ValueError:
    print("Invalid input\nPlease try again")
```
